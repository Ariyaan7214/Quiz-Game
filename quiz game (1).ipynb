```python
print("Welcome To Quiz Game!!")

print("IF YOUR SPELLING IS INCORRECT THEN THE IT IS CONSIDERED AS WRONG ANSWER!")

score=0
questionno=0

playing = input("Do You Want To Play? ").lower()

if playing=='yes':
    questionno += 1
    ques = input(f'\n{questionno}. what does CPU stand for? ').lower()
    
    if ques == 'central processing unit':
        score +=1
        print('correct! you got 1 point.')
        
    else:
        print('Incorrect!')
        print(f'correct answer is --> central processing unit')
    
    
    questionno += 1
    ques = input(f'\n{questionno}. what does GPU stand for? ').lower()
   
    if ques == 'graphical processing unit':
        score +=1
        print('correct! you got 1 point.')
        
    else:
        print('Incorrect!')
        print(f'correct answer is --> graphical processing unit')
        
    questionno += 1
    ques = input(f'\n{questionno}. what does RAM stand for? ').lower()
    
    if ques == 'random access memory':
        score +=1
        print('correct! you got 1 point.')
        
    else:
        print('Incorrect!')
        print(f'correct answer is --> random access memory')
    
    questionno += 1
    ques = input(f'\n{questionno}. what does PSU stand for? ').lower()
    
        
    if ques == 'power supply unit':
        score +=1
        print('correct! you got 1 point.')
        
    else:
        print('Incorrect!')
        print(f'correct answer is --> power supply unit')
    
    questionno += 1
    ques = input(f'\n{questionno}. what does ROM stand for? ').lower()
    
            
    if ques == 'read only memory':
        score +=1
        print('correct! you got 1 point.')
        
    else:
        print('Incorrect!')
        print(f'correct answer is --> read only memory')

else:
    print("Thank You! You Are Eliminated.")
    

    
print(f'Number Of Question Is {questionno}')

print(f'Your Score Is {score},Congratulations!')

print(f'You Got {score} out of {questionno}!')

percentage=(score*100)/questionno

print(f'{percentage}% Questions Are Correct.')
```
